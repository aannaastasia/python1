### Сортировка выбором

```python
def choice(arr):
    #Создаю копию массива, чтобы не изменять изначальный массив
    arrch = arr.copy()
    for k in range (m):
        for i in range (n-1):
            imin = i
            for j in range (i + 1, n):
                if arrch[k][j] < arrch[k][imin]: #Ищу минимальный элемент в строке
                    imin = j
            #Меняем местами минимальный элемент с i-ым элементом
            if i != imin:
                arrch[k][i], arrch[k][imin] = arrch[k][imin], arrch[k][i]
    return arrch
```

### Сортировка вставкой

```python
def insert(arr):
    #Создаю копию массива, чтобы не изменять изначальный массив
    arrin = arr.copy()
    for i in range(m):
        for j in range(n):
            temp = arrin[i][j]
            index = j
            while (temp < arrin[i][index-1]) and (index>0):
                arrin[i][index] = arrin[i][index-1]
                index -= 1
            arrin[i][index] = temp
    return arrin
```

### Сортировка обменом

```python
def exchange(arr):
    #Создаю копию массива, чтобы не изменять изначальный массив
    arrex = arr.copy()
    for i in range(m):
        for j in range(n-1):
            for k in range(n - j - 1):
                if arrex[i][k] > arrex[i][k+1]:
                    arrex[i][k], arrex[i][k + 1] = arrex[i][k + 1], arrex[i][k]
    return arrex
```

### Сортировка Шелла

```python
def shell(arr):
    #Создаю копию массива, чтобы не изменять изначальный массив
    array=arr.copy()
    for i in range(m):
        l= int(n / 2)
        while l > 0:
            for j in range(n):
                for h in range(int(j + l), n , l):
                    if array[i][j]>array[i][h]:
                        array[i][j], array[i][h] = array[i][h], array[i][j]
            l = int(l / 2)
    return array
```

### Турнирная сортировка

```python
def tournament(arr):
    #Создаю копию массива, чтобы не изменять изначальный массив
    arrt = arr.copy()
    for i in range(m):
        tournament_Sort(arrt[i])
    return arrt

def tournament_Sort(arrt):
    tree = [None] * 2 * (len(arrt) + len(arrt) % 2)
    index = len(tree) - len(arrt) - len(arrt) % 2
    
    #Сформируем турнирную таблицу
    for i, v in enumerate(arrt): #Возвращает пару (индекс, значение)
        tree[index + i] = (i, v)
    
    for j in range(len(arrt)):
        n = len(arrt)
        index = len(tree) - len(arrt) - len(arrt) % 2
        while index > -1:
            n = (n + 1) // 2
            for i in range(n):
                i = max(index + i * 2, 1) 
                if tree[i] != None and tree[i + 1] != None:
                    if tree[i][1] < tree[i + 1][1]:
                        tree[i // 2] = tree[i]
                    else:
                        tree[i // 2] = tree[i + 1]
                else:
                    tree[i // 2] = tree[i] if tree[i] != None else tree[i + 1]
            index -= n

        index, x = tree[0]
        arrt[j] = x #Минимум в турнирной таблице ставим на место j
        tree[len(tree) - len(arrt) - len(arrt) % 2 + index] = None
```

### Быстрая сортировка

```python
def fast(arr):
    #Создаю копию массива, чтобы не изменять изначальный массив
    mas = arr.copy()
    for i in range(m):
        fast_sort(mas[i])
    return mas
    
def fast_sort(arrf):
    if len(arrf) < 2:
        return arrf
    low, same, high = [], [], []
    t = arrf[random.randint(0, len(arrf) - 1)] #Выбирает рандомный элемент, чтобы взять за "середину"
    for item in arrf:
        if item < t:
            low.append(item)
        elif item == t:
            same.append(item)
        elif item > t:
            high.append(item)
    return fast_sort(low) + same + fast_sort(high) #рекурсивный вызов функции fast
```

### Пирамидальная сортировка

```python
def sort(arr, n, i):
    largest = i
    # индекс соседей (левого и правого)
    l = 2 * i + 1
    r = 2 * i + 2
    
    # Меняем соседей, если сосед больше
    if l < n and arr[i] < arr[l]:
        largest = l

    if r < n and arr[largest] < arr[r]:
        largest = r

    if largest != i:
        arr[i],arr[largest] = arr[largest],arr[i]
        # вызываем эту же функцию, но уже передаем индекс того, что мы поменяли
        sort(arr, n, largest)


def pyramid(array):
    arr=array.copy()
    for i in range(len(arr)):
        n = len(arr[i])
        
        # Идем с конца массива и вызываем sort для каждого элемента
        for j in range(n, -1, -1):
            sort(arr[i], n, j)

        # обмениваем 1 и последний (второй и предпоследний и т.д) элементы значениями и вызываем sort для 0 элемента
        for j in range(n-1, 0, -1):
            arr[i][j], arr[i][0] = arr[i][0], arr[i][j] 
            sort(arr[i], j, 0)
    return arr
```
